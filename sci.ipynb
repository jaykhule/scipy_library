```python
#scipy.integrate: For numerical integration (solving calculus problems).
from scipy import integrate

# Define a function to integrate
def func(x):
    return x**2

# Compute the integral of x^2 from 0 to 1
result, error = integrate.quad(func, 0, 1)
print("Integral result:", result)
print("Estimated error:", error)
```

```python
#Solving Linear Equations with scipy.linalg
#What is it?
#In engineering, systems of linear equations are common. SciPy provides tools to solve them efficiently.
#Example: Solve the equations:
#3x+y=9x+2y=8

from scipy.linalg import solve

# Coefficient matrix (A)
A = [[3, 1], [1, 2]]

# Constants (b)
b = [9, 8]

# Solve for x and y
x = solve(A, b)

print("Solution to the system of equations:", x)
```

```python
#Optimization with scipy.optimize
#What is it?
#Optimization finds the minimum or maximum value of a function, essential in engineering for design and analysis.
from scipy.optimize import minimize

# Define the function to minimize
def func(x):
    return (x - 3)**2 + 4

# Minimize the function
result = minimize(func, x0=0)  # Start with an initial guess x0 = 0

print("Optimal value of x:", result.x)
print("Minimum value of the function:", result.fun)

```

```python
#Interpolation with scipy.interpolate
#What is it?
#Interpolation estimates values between known data points
from scipy.interpolate import interp1d
import numpy as np

# Known data points
x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 1, 4, 9, 16])

# Create an interpolation function
f = interp1d(x, y, kind='linear')

# Interpolate at x = 2.5
y_interp = f(2.5)

print("Interpolated value at x=2.5:", y_interp)
```

```python
# Signal Processing with scipy.signal
from scipy.signal import butter, filtfilt
import numpy as np
import matplotlib.pyplot as plt

# Generate a noisy signal
t = np.linspace(0, 1, 500)
signal = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(500)

# Design a low-pass filter
b, a = butter(4, 0.1, btype='low')
filtered_signal = filtfilt(b, a, signal)

# Plot the original and filtered signals
plt.figure(figsize=(10, 5))
plt.plot(t, signal, label='Noisy Signal')
plt.plot(t, filtered_signal, label='Filtered Signal', linewidth=2)
plt.legend()
plt.show()
'''Explanation:
The butter function creates a low-pass filter to remove high-frequency noise.
filtfilt applies the filter to the noisy signal.
Result: The noisy signal is smoothed, making the sine wave more visible.'''
```

```python
# what if we want to count nonzeross,we can do this via count_nonzero()
import numpy as np
from scipy.sparse import csr_matrix
sh=np.array([[0,0,0],[0,0,1],[1,0,2]])
print(csr_matrix(sh).count_nonzero())   # give all count of all non zeros element
```

```python
# for removing the zeros elements from the matrix we will use eliminate_zeros()
import numpy as np
from scipy.sparse import csr_matrix
sh=np.array([[0,0,0],[0,0,1],[1,0,2]])
new= csr_matrix(sh)
new.eliminate_zeros()
print(new)
```

```python
import numpy as np
from scipy.sparse import csr_matrix

# Create a dense NumPy array with zeros and non-zero elements
dense_array = np.array([
    [0, 0, 3],
    [0, 0, 0],
    [4, 0, 5]
])

# Convert the dense array to a CSR sparse matrix
sparse_matrix = csr_matrix(dense_array)

# Before eliminating zeros
print("Sparse Matrix (Before eliminate_zeros):")
print(sparse_matrix)

# Eliminate explicit zeros
sparse_matrix.eliminate_zeros()

# After eliminating zeros
print("\nSparse Matrix (After eliminate_zeros):")
print(sparse_matrix)
```

```python
# eliminiting duplicate entries with the sum_duplicates() method:
import numpy as np
from scipy.sparse import csr_matrix
sh=np.array([[0,0,0],[0,0,1],[1,0,2]])
new= csr_matrix(sh)
new.sum_duplicates()
print(new)
```

```python
# to convert csr to csc with the tocsc():
import numpy as np
from scipy.sparse import csr_matrix
sh=np.array([[0,0,0],[0,0,1],[1,0,2]])
new= csr_matrix(sh).tocsc()
print(new)
```

```python
import numpy as np
from scipy.sparse.csgraph import connected_components
from scipy.sparse import csr_matrix
sh=np.array([[0,1,2],[1,0,0],[2,0,0]])
newsh=csr_matrix(sh)
print(connected_components(newsh))
```

```python
# Dijkstra : for finding shotest path in graph from one element to another
# it takes 3 args : return_predecessors , indices , limit
```

```python
# here we will find the shortedt path from element from 1 to 2 
import numpy as np
from scipy.sparse.csgraph import dijkstra
from scipy.sparse import csr_matrix
sh=np.array([[0,1,2],[1,0,0],[2,0,0]])
newsh=csr_matrix(sh)
print(dijkstra(newsh,return_predecessors=True,indices=0))
```

```python
# floyd warshall() method : it is for finding the shortest path between all the pair of element:
import numpy as np
from scipy.sparse.csgraph import floyd_warshall
from scipy.sparse import csr_matrix
sh=np.array([[0,1,2],[1,0,0],[2,0,0]])
newsh=csr_matrix(sh)
print(floyd_warshall(newsh,return_predecessors=True))
```
